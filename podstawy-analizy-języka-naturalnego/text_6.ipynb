````python
from spacy.lang.pl.examples import sentences
import spacy
nlp = spacy.load("pl_core_news_md")
````

````python
sentence = sentences[1]
doc = nlp(sentence)
````

# spaCy

### Stopwords

````python
nlp.Defaults.stop_words
````

````python
for token in doc:
    if token.is_alpha and not token.is_stop:
        print(token.text.lower())
````

### Lematyzacja

````python
for token in doc:
    if token.is_alpha:
        print(f"{token.text.lower():14} : {token.lemma_}")
````

### Części mowy

````python
for token in doc:
    print(f"{token.text.lower():14} : {token.lemma_:14} : {token.pos_:6} : {token.tag_}")
````
