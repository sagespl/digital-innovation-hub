````python
txt1 = pytesseract.image_to_string(img1)
````

````python
print(txt1)
````

````python
txt1_eng = pytesseract.image_to_string(img1, lang="eng")
````

````python
print(txt1_eng)
````

````python
txt1 == txt1_eng
````

````python
txt2 = pytesseract.image_to_string(img2)
````

````python
print(txt2)
````

````python
txt2_pol = pytesseract.image_to_string(img2, lang="pol")
````

````python
print(txt2_pol)
````

````python
print(pytesseract.image_to_boxes(img1))
````

````python
print(pytesseract.image_to_boxes(img2, lang="pol"))
````
